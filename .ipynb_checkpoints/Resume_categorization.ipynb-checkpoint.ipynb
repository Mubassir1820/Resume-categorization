# Data Preprocessing

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
import re
```

```python
stemmer = nltk.stem.porter.PorterStemmer()

def preprocess(text):
    ### Convert to lower case
    text = text.lower()
    
    ### Remove integers, punctuations with the help of Regular Expression
    text = re.sub('[^a-zA-Z]',' ', text)
    # Tokenization
    text = nltk.tokenize.word_tokenize(text)
    # Removing stop words
    text = [i for i in text if not i in nltk.corpus.stopwords.words('english')]
    # Stemming
    text = [stemmer.stem(i) for i in text]
    
    return ' '.join(text)
```

```python
df['Resume_str'] = df['Resume_str'].map(preprocess)
```

```python
# Word frequence

def wordfreq(df):
    count = df['Resume_str'].str.split(expand = True).stack().value_counts().reset_index()
    count.columns = ['Word', 'Frequency']
    
    return count.head()
```

# Data Cleaning

```python
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Instantiate stopwords
stop_words = stopwords.words('english')
```

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
```

```python
# to remove stop words and words that are less than 2 lengths
def remove_stop_words(text):
    result = []
    
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3 and token not in stop_words:
            result.append(token)
            
    return result
```

```python
df['Resume'] = df['Resume_str'].apply(remove_stop_words).astype(str)
```

```python
df['Resume'][0]
```

```python
# Train_test_splitting
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['Resume'], df['Category'], test_size = 0.2)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.2)
```

```python
# Vectorization of the corpus 
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
countvectorizer_train = vectorizer.fit_transform(X_train).astype(float)
countvectorizer_test = vectorizer.transform(X_val).astype(float)
```

# Model Training

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

RFC = RandomForestClassifier(random_state = 42)

param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}

grid = GridSearchCV(estimator=RFC, param_grid=param_grid, cv= 5, scoring='accuracy', return_train_score=False, verbose=1)
grid_search = grid.fit(countvectorizer_train, y_train)

print(grid_search.best_params_)
```

```python
RF_model = RandomForestClassifier(random_state = 42, max_features = 'auto', n_estimators = 500, max_depth = 8, criterion = 'gini')
RF_model.fit(countvectorizer_train, y_train)
```

```python
# prediction
prediction = RF_model.predict(countvectorizer_test)
prediction
```

# Evaluation

```python
print("training Score: {:.2f}".format(RF_model.score(countvectorizer_train, y_train)))
print("test Score: {:.2f}".format(RF_model.score(countvectorizer_test, y_val)))
```

```python
# Accuracy
from sklearn.metrics import accuracy_score   

score = accuracy_score(y_val, prediction)
print(f'The accuracy of the model is: {score}')
```

```python
from sklearn import metrics
print("model report: %s: \n %s\n" % (RF_model, metrics.classification_report(y_val, prediction)))
```

```python
# Visualization of classification report
import seaborn as sns
from sklearn.metrics import classification_report

clf_report = classification_report(y_val,
                                   prediction,
                                   output_dict=True)

sns.heatmap(pd.DataFrame(clf_report).iloc[:-1, :].T, annot=True)
```
